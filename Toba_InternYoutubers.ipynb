```python
# Scatter plot
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Suscribers', y='Likes', data=df)
plt.xlabel('Suscribers')
plt.ylabel('Likes')
plt.title('Scatter Plot: Suscribers vs Likes')
plt.show()
```

```python
The statement "Correlation between Subscribers and Likes: 0.21163868368008348" indicates the strength and direction of the linear relationship between the number of subscribers and the number of likes for a particular dataset. The value 0.21163868368008348 is the correlation coefficient.

The correlation coefficient is a numerical measure that ranges from -1 to 1. Here's how to interpret it:

Strength of the Relationship:

A correlation coefficient of 1 indicates a perfect positive linear relationship, meaning as one variable (subscribers) increases, the other variable (likes) also increases proportionally.
A correlation coefficient of -1 indicates a perfect negative linear relationship, meaning as one variable increases, the other decreases proportionally.
A correlation coefficient of 0 suggests no linear relationship between the variables.
Direction of the Relationship:

A positive correlation coefficient (closer to 1) implies that as the number of subscribers increases, the number of likes also tends to increase.
A negative correlation coefficient (closer to -1) implies that as the number of subscribers increases, the number of likes tends to decrease.
In your case, the correlation coefficient is positive (0.21163868368008348), indicating a positive but relatively weak linear relationship between the number of subscribers and the number of likes. This suggests that there is some tendency for videos with more subscribers to also have more likes, but the relationship is not very strong.

It's important to note that correlation does not imply causation. Even if there is a correlation between subscribers and likes, it doesn't necessarily mean that one variable is causing the other to change. Other factors may be influencing both variables, or the relationship may be coincidental.
```

6. Brands and Collaborations:
- Analyze whether streamers with high performance metrics receive more brand collaborations and
marketing campaigns.

```python
high_suscribers_threshold = df['Suscribers'].quantile(0.75, interpolation="nearest")
high_likes_threshold = df['Likes'].quantile(0.75, interpolation="nearest")
high_comments_threshold = df['Comments'].quantile(0.75, interpolation="nearest")
```

```python
df['Highperformance'] = ((df['Suscribers'] >= high_suscribers_threshold) &
                          (df['Likes'] >= high_likes_threshold) &
                           (df['Comments'] >= high_comments_threshold))
```

```python
df['Engagement'] = df['Likes'] + df['Comments']
high_performance_engagement = df.loc[df['Highperformance']]['Engagement'].sum()
low_performance_engagement = df.loc[~df['Highperformance']]['Engagement'].sum()
```

```python
# Plot the distribution of engagement for high and low performance
plt.figure(figsize=(6, 4))
sns.barplot(x=['High Performance', 'Low Performance'], y=[high_performance_engagement, low_performance_engagement], palette='viridis')
plt.xlabel('Performance Category')
plt.ylabel('Total Engagement')
plt.title('Distribution of Engagement for High and Low Performance')
plt.show()
```

7. Benchmarking:
- Identify streamers with above-average performance in terms of subscribers, visits, likes, and comments.
- Who are the top-performing content creators?

```python
average_metrics = df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
```

```python
above_average_streamers = df[
    (df['Suscribers'] > average_metrics['Suscribers']) &
    (df['Visits'] > average_metrics['Visits']) &
    (df['Likes'] > average_metrics['Likes']) &
    (df['Comments'] > average_metrics['Comments'])]
```

```python
top_creators = df.nlargest(10, 'Suscribers')
```

```python
print("Above-Average Streamers:")
print(above_average_streamers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']])
print("\nTop-Performing Content Creators:")
print(top_creators[['Username', 'Suscribers']])
```

```python
user_item_matrix = df.pivot_table(index='Username', columns='Categories', values='Suscribers', fill_value=0)
```

```python
user_item_matrix
```

```python
pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(user_item_matrix)
```

```python
def get_recommendations(username, cosine_sim=cosine_sim):
    idx = user_item_matrix.index.get_loc(username)
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]  # Exclude the user itself, which will have a similarity score of 1
    streamer_indices = [i[0] for i in sim_scores]
    return user_item_matrix.index[streamer_indices]
```

```python
recommended_streamers = get_recommendations('tseries')
```
